#### Notes
* For NRMSE (Normalized Root Mean Square Error) I used the range of highest and lowest values in data i.e.
\begin{align}
NMRSE = \frac{1}{y_{max}-y_{min}} \sqrt{\frac{\sum\limits_{i}^{n} (y_i - \hat{y})^2}{n}}
\end{align}

## 1. Data Cleaning and Processing

```python
df_movie = pd.read_csv("moviesDB.csv")
df_movie.dropna(inplace = True)
```

```python
df_movie.genre.unique()
```

A few of these genres only have a few films in the database. Very small in comparison to the actual size of the data set. So I will drop them.

```python
df_movie = df_movie[df_movie['genre']!='Western']
df_movie = df_movie[df_movie['genre']!='Thriller']
df_movie = df_movie[df_movie['genre']!='Romance']
df_movie = df_movie[df_movie['genre']!='Sci-Fi']
df_movie = df_movie[df_movie['genre']!='Family']
```

```python
df_movie.rating.unique()
```

Same for ratings

```python
df_movie = df_movie[df_movie['rating']!='X']
df_movie = df_movie[df_movie['rating']!='Approved']
df_movie = df_movie[df_movie['rating']!='TV-MA']
```

I want to change the 'Release' category to only include the release month

```python
df_movie = df_movie.rename(columns={'released': 'Month'})
df_movie['Month'] = df_movie['Month'].str.split(' ').str[0]
```

```python
df_movie.Month.unique()
```

```python
df_movie = df_movie[df_movie['Month']!='1982']
df_movie = df_movie[df_movie['Month']!='1985']
df_movie = df_movie[df_movie['Month']!='1995']
df_movie = df_movie[df_movie['Month']!='2019']
```

```python
df_movie = df_movie.set_index(np.arange(0, df_movie.shape[0], dtype=int))
df_movie
```

This is good for analysis and visualization, but these categorical variables need to be numerical to train a model. So, I'll make a second dataset, $\verb"df_train"$, to train our model.

```python
df_train = pd.concat([df_movie.loc[:,'name'], pd.get_dummies(df_movie.rating, dtype=int), pd.get_dummies(df_movie.genre, dtype=int), df_movie.loc[:,'year':]], ignore_index=False, sort=False, axis=1)
```

First, I'll replace the month string value with the number.

```python
Calender = {"January":1, "February":2, "March":3, "April":4, "May":5, "June":6, "July":7, "August":8, "September":9, "October":10, "November":11, "December":12}

for month in Calender:
    df_train['Month'] = df_train['Month'].replace(month, Calender[month])
```

Next I want to convert the countries to a catagorical value

```python
df_train.country = pd.Categorical(df_movie.country)
df_train['Country Codes'] = df_train.country.cat.codes
```

I won't train the model with the 'company', 'country', 'star', 'writer' or 'director' category because we can't numerically encode them. (I want to return to this eventually and try to find a way to numerically encode them.)

```python
df_train = df_train.drop(columns=['company', 'country', 'star', 'writer', 'director'], axis=1)
df_train
```

#### 1. Distribution of Ratings

```python
max_score = 9.
min_score = 4.

score_std = np.std(df_movie['score'])
score_mean = np.mean(df_movie.loc[:,['score']])
film_total = df_movie.loc[:,['score']].shape[0]
```

```python
films_per_rating = df_movie.loc[:,['score']].groupby('score').size()

bin_width = 0.1

gauss = (bin_width*film_total/np.sqrt(2*np.pi*score_std**2))*np.exp(-0.5*((np.arange(0,10.1,0.1)-score_mean)/score_std**2)**2)

fig = plt.figure()

fig.set_figwidth(10)
fig.set_figheight(3)

plt.bar(x = films_per_rating.keys(), height = films_per_rating, width=0.1, label = 'Data')
plt.plot(np.arange(0,10.1,0.1), gauss, color = 'r', label = 'Gaussian')
plt.xticks(np.arange(0,10.1,0.5))
plt.yticks(np.arange(0,261,20))
plt.xlim(1.5,9.5)
plt.xlabel('Score')
plt.ylabel('# of Films')
plt.legend(loc="upper left")
plt.title('1. Distribution of Ratings')


plt.show()
```

```python
smooth_factor = 10
bin_num = 100
```

```python
df_movie['profit margin'] = df_movie['gross']/df_movie['budget']


rating_by_gross = df_movie.loc[:,['gross','score']].groupby('gross').mean().reset_index()
rating_by_gross = rating_by_gross.groupby(np.arange(len(rating_by_gross))//smooth_factor).mean()
rating_by_budget = df_movie.loc[:,['budget','score']].groupby('budget').mean().reset_index()
rating_by_budget = rating_by_budget.groupby(np.arange(len(rating_by_budget))//smooth_factor).mean()
rating_by_profitm = df_movie.loc[:,['profit margin','score']].groupby('profit margin').mean().reset_index()
rating_by_profitm = rating_by_profitm.groupby(np.arange(len(rating_by_profitm))//smooth_factor).mean()

films_per_gross = df_movie.loc[:,['gross']]
films_per_budget = df_movie.loc[:,['budget']]
films_per_profitm = df_movie.loc[:,['profit margin']]

logbins_gross = np.logspace(np.log10(films_per_gross['gross'].min()),np.log10(films_per_gross['gross'].max()),bin_num)
logbins_budget = np.logspace(np.log10(films_per_budget['budget'].min()),np.log10(films_per_budget['budget'].max()),bin_num)
logbins_profitm = np.logspace(np.log10(films_per_profitm['profit margin'].min()),np.log10(films_per_profitm['profit margin'].max()),bin_num)


fig = plt.figure()

fig.set_figwidth(12)
fig.set_figheight(12)

plt.subplots_adjust(wspace=0.3, hspace=0.4)

fig.suptitle('Genre Distribution', fontsize=16)


ax = fig.add_subplot(3, 2, 1)
plt.semilogx(rating_by_gross['gross'], rating_by_gross['score'], label = 'Data')
plt.plot(rating_by_gross['gross'], np.full(len(rating_by_gross['gross']), score_mean), color='r', label = 'Mean')
plt.plot(rating_by_gross['gross'], np.full(len(rating_by_gross['gross']), score_mean-score_std), color='r', linestyle='dotted', label = 'Std. Dev.')
plt.plot(rating_by_gross['gross'], np.full(len(rating_by_gross['gross']), score_mean+score_std), color='r', linestyle='dotted')
plt.xlim(5e+3, 5e+9)
plt.ylim(min_score,max_score)
plt.legend(loc="upper left")
plt.xlabel('Gross Profit')
plt.ylabel('IMDB Score')
plt.title('9a. Score Distribution by Gross Profit')

ax = fig.add_subplot(3, 2, 2)
plt.hist(films_per_gross['gross'], bins=logbins_gross)
plt.xscale('log')
plt.xlim(5e+3, 5e+9)
plt.xlabel('Gross Profit')
plt.ylabel('# of Films')
plt.title('9d. Distribution of Gross Profit')

ax = fig.add_subplot(3, 2, 3)
plt.semilogx(rating_by_budget['budget'], rating_by_budget['score'], label = 'Data')
plt.plot(rating_by_gross['gross'], np.full(len(rating_by_gross['gross']), score_mean), color='r', label = 'Mean')
plt.plot(rating_by_gross['gross'], np.full(len(rating_by_gross['gross']), score_mean-score_std), color='r', linestyle='dotted', label = 'Std. Dev.')
plt.plot(rating_by_gross['gross'], np.full(len(rating_by_gross['gross']), score_mean+score_std), color='r', linestyle='dotted')
plt.xlim(5e+3, 5e+9)
plt.ylim(min_score,max_score)
plt.legend(loc="upper left")
plt.xlabel('Budget')
plt.ylabel('IMDB Score')
plt.title('9b. Score Distribution by Budget')


ax = fig.add_subplot(3, 2, 4)
plt.hist(films_per_budget['budget'], bins=logbins_budget)
plt.xscale('log')
plt.xlim(5e+3, 5e+9)
plt.xlabel('Budget')
plt.ylabel('# of Films')
plt.title('9e. Distribution of Budget')

ax = fig.add_subplot(3, 2, 5)
plt.semilogx(rating_by_profitm['profit margin'], rating_by_profitm['score'], label = 'Data')
plt.plot(rating_by_profitm['profit margin'], np.full(len(rating_by_profitm['profit margin']), score_mean), color='r', label = 'Mean')
plt.plot(rating_by_profitm['profit margin'], np.full(len(rating_by_profitm['profit margin']), score_mean-score_std), color='r', linestyle='dotted', label = 'Std. Dev.')
plt.plot(rating_by_profitm['profit margin'], np.full(len(rating_by_profitm['profit margin']), score_mean+score_std), color='r', linestyle='dotted')
plt.xlim(1e-2, 1e+2)
plt.ylim(min_score,max_score)
plt.legend(loc="upper left")
plt.xlabel('Gross Profit/Budget')
plt.ylabel('IMDB Score')
plt.title('9c. Score Distribution by Gross Profit: Budget')

ax = fig.add_subplot(3, 2, 6)
plt.hist(films_per_profitm['profit margin'], bins=logbins_profitm)
plt.xscale('log')
plt.xlim(1e-2, 1e+2)
plt.xlabel('Gross Profit/Budget')
plt.ylabel('# of Films')
plt.title('9f. Distribution of Gross Profit: Budget')

plt.show()
```

When deciding which parameters to train the model with, I trained the model with few different sets of parameters. Exculding some and leaving others in changed the NRMSE (Normalized Root Mean Squared Error) and R^2 values only slightly, with some favoring NRMSE and vice versa. Ultimately I favored making the NMRSE as low as possible. The set of parameters that made the NMRSE as small as possible was leaving them all in the model, so I'll proceed with that.

We will make some common functions for later

```python
def graph(w, predictions, y_test_set, rate_std, range_min, range_max):
    """
    This creates standardized graphs to analyze the machine learning models we used.
    
    w - The width of the bars used in the bar graphs
    predictions - The set of scores predicted by the model from the test data
    y_test_set - The correct scores of the test data
    rate_std - The standard deviation of ratings
    range_min - The smallest index we want to plot
    range_max - The largest index we want to plot
    """
    x = np.arange(0,len(predictions))

    fig = plt.figure()
    
    fig.set_figwidth(10)
    fig.set_figheight(8)

    ax = fig.add_subplot(3, 1, 1)
    ax.bar(x-w/2, y_test_set, width=w, color='b', align='center')
    ax.bar(x+w/2, predictions, width=w, color='r', align='center')
    plt.xlim(range_min,range_max)


    ax = fig.add_subplot(3, 1, 2)
    ax.bar(x, y_test_set - predictions, width=w, color='b', align='center')
    plt.xlim(range_min,range_max)

    ax = fig.add_subplot(3, 1, 3)
    ax.bar(x, abs(y_test_set - predictions), width=w, color='b', align='center')
    ax.plot(x, np.full(len(x), rate_std), color='r')
    plt.xlim(range_min,range_max)
    plt.show()
    
    
def Metric_Report(predictions, y_test_set):
    """
    This creates a table of some useful statistics to evaluate our models.
    
    predictions - The set of scores predicted by the model from the test data
    y_test_set - The correct scores of the test data
    """
    NMAE = np.mean(np.absolute(predictions - y_test_set))/(max(y_test_set)-min(y_test_set)) # Normalized Mean Absolute Error
    NRMSE = np.sqrt(np.mean((predictions - y_test_set) ** 2))/(max(y_test_set)-min(y_test_set)) # Normalized Root Mean Squared Error
    R2 = r2_score(y_test_set, predictions) # R squared value
    return pd.DataFrame([{'NMAE':NMAE, 'NRMSE':NRMSE, 'R2':R2}])
```
